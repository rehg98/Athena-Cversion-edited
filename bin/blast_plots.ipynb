## Read and plot history dump: KE vs. t and KE/TotE vs. t

```python
hst = ath_hst.read("Blast_Test.hst")
```

```python
time = hst['time']
KE = hst['x1KE'] + hst['x2KE'] + hst['x3KE']
internalE = hst['internal_E']
KEoverTot = KE / (KE + internalE)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 4))
fig.subplots_adjust(wspace=0.3)

ax[0].plot(time, KE)
ax[0].set_title("Bubble Kinetic Energy vs. Time")
ax[0].set_xlabel("Time")
ax[0].set_ylabel("Kinetic Energy")

# y-value should converge to ~0.283
ax[1].plot(time, KEoverTot)
ax[1].set_title("Bubble Kinetic Energy Fraction vs. Time")
ax[1].set_xlabel("Time")
ax[1].set_ylabel("Kinetic Energy / Total Energy")
ax[1].axhline(0.283,ls=':')

plt.show()
```

## Read density, pressure, and velocity data from VTK files

```python
# Set up cell-centered coordinates

ds=pa.AthenaDataSet('Blast_Test.0001.vtk')
xcc,ycc,zcc=pa.cc_arr(ds.domain)
rcc=np.sqrt(xcc[np.newaxis,np.newaxis,:]**2+ycc[np.newaxis,:,np.newaxis]**2+zcc[:,np.newaxis,np.newaxis]**2)

# domain information
xmin=ds.domain['left_edge']
xmax=ds.domain['right_edge']
dx=ds.domain['dx']
Nx=ds.domain['Nx']

# center of the domain
icen=Nx/2

# this sets up for image plots based on the domain physical 
xyextent=[xmin[0],xmax[0],xmin[1],xmax[1]]
xzextent=[xmin[0],xmax[0],xmin[2],xmax[2]]
yzextent=[xmin[1],xmax[1],xmin[2],xmax[2]]
```
